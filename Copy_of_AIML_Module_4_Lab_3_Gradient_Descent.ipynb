## Lab 3: Gradient Descent

Gradient descent is a very important algorithm to understand, as it underpins many of the more advanced algorithms used in Machine Learning and Deep Learning.

A brief overview of the algorithm is

*   start with a random initialization of the solution.
*   incrementally change the solution by moving in the direction of negative gradient of the objective function.
*   repeat the previous step until some convergence criteria is met.

The key equation for change in weight is:
$$w^{k+1} \leftarrow w^k - \eta \Delta J$$

In this lab, we will discuss stochastic gradient descent, mini-batch gradient descent and batch gradient descent.

## Cost Function

The equation for calculating cost function is as shown below. The cost function is only for linear regression. For other algorithms, the cost function will be different and the gradients would have to be derived from the cost functions

\begin{equation}
J(y_{pred}) = \frac{1}{2} m \sum_{i=1}^{m} (h(y_{pred})^{(i)} - y^{(i)})^2
\end{equation}

```python
def cal_cost(y_pred, X, y):
    '''
    Calculates the cost for given X and Y.
    y_pred = Vector of y_preds
    X = Row of X's np.zeros((2, j))
    y = Actual y's np.zeros((2, 1))

    where:
        j is the no of features
    '''

    m = len(y)

    predictions = X.dot(y_pred)
    cost = (1 / 2 * m) * np.sum(np.square(predictions - y))

    return cost
```

## Gradients

\begin{equation}
y_{pred_0}: = y_{pred_0} -\alpha . (1/m .\sum_{i=1}^{m}(h(y_{pred}^{(i)} - y^{(i)}).X_0^{(i)})
\end{equation}
\begin{equation}
y_{pred_1}: = y_{pred_1} -\alpha . (1/m .\sum_{i=1}^{m}(h(y_{pred}^{(i)} - y^{(i)}).X_0^{(i)})
\end{equation}

\begin{equation}
y_{pred_j}: = y_{pred_j} -\alpha . (1/m .\sum_{i=1}^{m}(h(y_{pred}^{(i)} - y^{(i)}).X_0^{(i)})
\end{equation}

```python
def gradient_descent(X, y, y_pred, learning_rate=0.01, iterations=100):
    '''
    X = Matrix of X with added bias units
    y = Vector of Y
    y_pred = Vector of y_preds np.random.randn(j, 1)
    learning_rate
    iterations = no of iterations

    Returns the final y_pred vector and array of cost history over no of iterations
    '''

    m = len(y)
    cost_history = np.zeros(iterations)
    y_pred_history = np.zeros((iterations, 2))

    for it in range(iterations):
        prediction = np.dot(X, y_pred)
        y_pred = y_pred - (1 / m) * learning_rate * (X.T.dot((prediction - y)))
        y_pred_history[it,:] = y_pred.T
        cost_history[it]  = cal_cost(y_pred, X, y)

    return y_pred, cost_history, y_pred_history
```

# Stochastic Gradient Descent

In Batch Gradient Descent we were considering all the examples for every step of Gradient Descent. But what if our dataset is very huge. Deep learning models crave for data. The more the data the more chances of a model to be good. Suppose our dataset has 5 million examples, then just to take one step the model will have to calculate the gradients of all the 5 million examples. This does not seem an efficient way. To tackle this problem we have Stochastic Gradient Descent. In Stochastic Gradient Descent (SGD), we consider just one example at a time to take a single step.

```python
def stocashtic_gradient_descent(X, y, y_pred, learning_rate=0.01, iterations=10):
    '''
    X = Matrix of X with added bias units
    y = Vector of Y
    y_pred = Vector of y_pred np.random.randn(j,1)
    learning_rate
    iterations = no of iterations

    Returns the final y_pred vector and array of cost history over no of iterations
    '''

    m = len(y)
    cost_history = np.zeros(iterations)

    for it in range(iterations):
        cost = 0.0

        for i in range(m):
            rand_ind = np.random.randint(0,m)
            X_i = X[rand_ind, :].reshape(1, X.shape[1])
            y_i = y[rand_ind].reshape(1,1)
            prediction = np.dot(X_i, y_pred)

            y_pred = y_pred - (1 / m) * learning_rate *(X_i.T.dot((prediction - y_i)))
            cost += cal_cost(y_pred, X_i, y_i)

        cost_history[it]  = cost

    return y_pred, cost_history
```

```python
lr = 0.5
n_iter = 50
y_pred = np.random.randn(2, 1)
X_b = np.c_[np.ones((len(X), 1)), X]
y_pred, cost_history = stocashtic_gradient_descent(X_b, y, y_pred, lr, n_iter)

print('y_pred[0]: {:0.3f}\ny_pred[1]: {:0.3f}'.format(y_pred[0][0], y_pred[1][0]))
print('Final error: {:0.3f}'.format(cost_history[-1]))
```

# Mini Batch Gradient Descent

We have seen the Batch Gradient Descent. We have also seen the Stochastic Gradient Descent. Batch Gradient Descent can be used for smoother curves. SGD can be used when the dataset is large. Batch Gradient Descent converges directly to minima. SGD converges faster for larger datasets. But, since in SGD we use only one example at a time, we cannot implement the vectorized implementation on it. This can slow down the computations. To tackle this problem, a mixture of Batch Gradient Descent and SGD is used.
Neither we use all the dataset all at once nor we use the single example at a time. We use a batch of a fixed number of training examples which is less than the actual dataset and call it a mini-batch. Doing this helps us achieve the advantages of both the former variants we saw.

```python
def minibatch_gradient_descent(X, y, y_pred, learning_rate=0.01, iterations=10, batch_size=20):
    '''
    X = Matrix of X without added bias units
    y = Vector of Y
    y_pred = Vector of y_preds np.random.randn(j, 1)
    learning_rate
    iterations = no of iterations

    Returns the final theta vector and array of cost history over no of iterations
    '''

    m = len(y)
    cost_history = np.zeros(iterations)
    n_batches = int(m / batch_size)

    for it in range(iterations):
        cost = 0.0
        indices = np.random.permutation(m)
        X = X[indices]
        y = y[indices]

        for i in range(0, m, batch_size):
            X_i = X[i: i + batch_size]
            y_i = y[i: i + batch_size]

            X_i = np.c_[np.ones(len(X_i)), X_i]
            prediction = np.dot(X_i, y_pred)

            y_pred = y_pred - (1 / m) * learning_rate * (X_i.T.dot((prediction - y_i)))
            cost += cal_cost(y_pred, X_i, y_i)

        cost_history[it]  = cost

    return y_pred, cost_history
```

```python
lr = 0.1
n_iter = 200
y_pred = np.random.randn(2,1)
y_pred, cost_history = minibatch_gradient_descent(X, y, y_pred, lr, n_iter)

print('y_pred[0]: {:0.3f}\ny_pred[1]: {:0.3f}'.format(y_pred[0][0], y_pred[1][0]))
print('Final error: {:0.3f}'.format(cost_history[-1]))
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np

# Load Iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2]  # Selecting the first two features for simplicity
y = (iris.target != 0) * 1  # Convert to binary classification (1 for versicolor, 0 for setosa)

# Standardize features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Add a bias term to the features
X_bias = np.c_[np.ones((X.shape[0], 1)), X]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_bias, y, test_size=0.2, random_state=42)

# Linear Regression Model
class LinearRegression:
    def __init__(self, learning_rate=0.01, epochs=1000):
        self.learning_rate = learning_rate
        self.epochs = epochs

    def fit(self, X, y):
        self.theta = np.zeros(X.shape[1])
        m = len(y)

        for epoch in range(self.epochs):
            gradients = 2/m * X.T.dot(X.dot(self.theta) - y)
            self.theta -= self.learning_rate * gradients

    def predict(self, X):
        return X.dot(self.theta)

# Function to evaluate the model
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    predictions_train = model.predict(X_train)
    predictions_test = model.predict(X_test)

    # You can add evaluation metrics here (e.g., accuracy, precision, recall, etc.)

    return predictions_train, predictions_test

# Batch Gradient Descent
lr_batch = LinearRegression(learning_rate=0.01, epochs=1000)
predictions_train_batch, predictions_test_batch = evaluate_model(lr_batch, X_train, y_train, X_test, y_test)

# Stochastic Gradient Descent
lr_sgd = LinearRegression(learning_rate=0.01, epochs=1000)
predictions_train_sgd, predictions_test_sgd = evaluate_model(lr_sgd, X_train, y_train, X_test, y_test)

# Mini-batch Gradient Descent
lr_mini_batch = LinearRegression(learning_rate=0.01, epochs=1000)
predictions_train_mini_batch, predictions_test_mini_batch = evaluate_model(lr_mini_batch, X_train, y_train, X_test, y_test)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Function to plot decision boundary
def plot_decision_boundary(model, X, y, title):
    h = .02  # step size in the mesh

    x_min, x_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    y_min, y_max = X[:, 2].min() - 1, X[:, 2].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[np.ones((len(xx.ravel()), 1)), xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, cmap=ListedColormap(['#FFAAAA', '#AAAAFF']))
    plt.scatter(X[:, 1], X[:, 2], c=y, cmap=ListedColormap(['#FF0000', '#0000FF']), edgecolors='k', marker='o')
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

# Plot decision boundary for Batch Gradient Descent
plot_decision_boundary(lr_batch, X_train, y_train, "Batch Gradient Descent")

# Plot decision boundary for Stochastic Gradient Descent
plot_decision_boundary(lr_sgd, X_train, y_train, "Stochastic Gradient Descent")

# Plot decision boundary for Mini-batch Gradient Descent
plot_decision_boundary(lr_mini_batch, X_train, y_train, "Mini-batch Gradient Descent")

# Display results
def display_results(predictions_train, predictions_test, title):
    print(f"\nResults for {title}:\n")
    print("Train Set Predictions:")
    print(predictions_train)
    print("Test Set Predictions:")
    print(predictions_test)

# Display results for Batch Gradient Descent
display_results(predictions_train_batch, predictions_test_batch, "Batch Gradient Descent")

# Display results for Stochastic Gradient Descent
display_results(predictions_train_sgd, predictions_test_sgd, "Stochastic Gradient Descent")

# Display results for Mini-batch Gradient Descent
display_results(predictions_train_mini_batch, predictions_test_mini_batch, "Mini-batch Gradient Descent")
```

3Q )Compare the effects of changing learning rate by the same amount in Batch GD, SGD and Mini-batch GD.

- **Batch Gradient Descent (BGD):**
  - Utilizes the entire dataset for parameter updates.
  - Learning rate adjustments directly affect the step size towards the minimum.
  - Higher learning rates lead to faster convergence but risk overshooting, while lower rates result in steadier convergence with less chance of overshooting.

- **Stochastic Gradient Descent (SGD):**
  - Updates parameters using one random training example at a time.
  - Learning rate changes influence the size of adjustments after each observed example.
  - Higher learning rates accelerate convergence but introduce erratic steps and potential instability, whereas lower rates offer stability at the cost of slower convergence.

- **Mini-batch Gradient Descent (MBGD):**
  - Uses small batches of training examples, balancing BGD's stability and SGD's erratic behavior.
  - Adjusting the learning rate impacts the step size during updates in each mini-batch.
  - Higher learning rates speed up convergence but may lead to oscillations or instability, while lower rates provide smoother convergence at the expense of potentially requiring more iterations.

- **Effects of Learning Rate Changes:**
  - **Speed of Convergence:**
    - Higher rates accelerate convergence but pose a risk of overshooting.
    - Lower rates ensure stability but might result in slower convergence.
  - **Smoothness of Convergence:**
    - Higher rates lead to erratic steps and potential instability.
    - Lower rates offer steadier convergence but may take longer.
  - **Overall Performance:**
    - Learning rate choice directly impacts final model performance and its ability to generalize to new data.
    - Finding the ideal rate involves a balance between speed and stability for optimal generalization.
